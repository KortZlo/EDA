**Задание 3**

В игре Plants & Gardens каждый месяц проводятся тематические события, ограниченные по времени. В них игроки могут получить уникальные предметы для сада и персонажей, дополнительные монеты или бонусы. Для получения награды требуется пройти ряд уровней за определенное время. С помощью каких метрик можно оценить результаты последнего прошедшего события?

Предположим, в другом событии мы усложнили механику событий так, что при каждой неудачной попытке выполнения уровня игрок будет откатываться на несколько уровней назад. Изменится ли набор метрик оценки результата? Если да, то как?

**Ответ**

*Событие 1*

В моем понимании любой игровой ивент преследует 3 основных цели: 
1. Привлечение новой аудитории; 
2. Удержание и/или возврат старых игроков; 
3. Получение прибыли от транзакций. 

Поэтому основные KPI будут следующие: 
1. Install или New Users - для оценки новой аудитории;
2. Retention + Lifetime + DAU, WAU, MAU + Sessions\Average Session Duration - для работы с основной аудиторией; 
3. ATV + ARPU/ARPPU - для подсчета средней стоимости покупок за период, среднего чека + cредний доход с одного пользователя/платящего пользователя.


Для оценки события так же нужно будет отслеживать: 
- Event Success Rate - % игроков, успешно прошедших событие. Не должно быть слишком легко или слишком сложно(если это не было так задумано). Нужны ли бустеры (если таковые имеются по механике игры) и т.д.;
- Churn Rate - как изменился отток клиентов;
- Selected Award - метрика, отслеживающая частоту выбора той или иной награды/предмета, если предусмотрена механика выбора.

Группа метрик для понимания правильности дизайна уровней и баланса. Механик монетизации и буста игроков.

- Average Event Level - средний уровень, пройденный игроками;
- Time per Level - затраченое время на прохождение опредленного уровня;
- Attempts per Level - количество попыток прохождения уровня. 

*Событие 2*

Основные KPI оставим прежними, чтобы иметь возможность сравнить два события.
Добавим две новые метрики:
- Avg Level Rollback - средний уровень отката игроков;
- Level Rollback per User - количество отката уровней для каждого игрока.
